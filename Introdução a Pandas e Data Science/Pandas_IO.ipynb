# Incluindo ID dos alunos

```python
nomes['email'] = nomes.nome.str.cat(nomes.dominio).str.lower()
nomes
```

# Matriculando os alunos nos cursos

```python
#nomes['matriculas'] = np.random.choice(a = list(range(1,7)),size = len(nomes),p=[0.5,0.25,0.12,0.07,0.04,0.02])
nomes['matriculas'] = np.ceil(np.random.exponential(size=len(nomes))*1.5).astype(int)
```

# Selecionando Cursos

```python
todas_matriculas = []
x = np.random.rand(20)
prob = x/sum(x)
```

```python
for index,row in nomes.iterrows():
  id = row.id_aluno
  matriculas = row.matriculas
  for i in range(matriculas):
    todas_matriculas.append([id,np.random.choice(cursos.index,p = prob)]) 
matriculas = pd.DataFrame(todas_matriculas,columns=['id_aluno','id_curso'])
matriculas
```

```python
matricula_por_curso = matriculas.groupby('id_curso').count().join(cursos['Nome_do_Curso']).rename(columns={'id_aluno':'quantidade_de_alunos'})
```
