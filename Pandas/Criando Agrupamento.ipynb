## Criando Agrupamento

```python
import pandas as pd
```

```python
dados=pd.read_csv('data/aluguel_residencial.csv',sep=';')
```

```python
dados.head(10)
```

```python
dados['Valor'].mean()
```

```python
dados['Bairro'].unique()
```

```python
{x:dados[dados['Bairro']==x]['Valor'].mean() for x in dados['Bairro'].unique()}
```

```python
dados.groupby('Bairro')['Valor'].mean().round(2)
```

## Estatísticas Descritivas

```python
dados.groupby('Bairro')['Valor'].describe().sort_values(by='count',ascending = False)
```

```python
dados.groupby('Bairro')['Valor'].aggregate(['min','max']).rename(columns={'min': 'Mínimo','max':'Máximo'})
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
plt.rc('figure',figsize=(20,10))
```

```python
ax = dados.groupby('Bairro')['Valor'].std().plot.bar(color='blue')
ax.set_ylabel('Desvio do Valor do Aluguel')
ax.set_title('Desvio Padrão do Valor do Aluguel por Bairro',{'fontsize':22})
```
